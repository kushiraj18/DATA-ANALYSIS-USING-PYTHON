```python
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data['sex'] = data['sex'].map({'female': 0, 'male': 1})
data['smoker'] = data['smoker'].map({'no': 0, 'yes': 1})
```

```python
data = pd.get_dummies(data, columns=['region'], drop_first=True)
```

```python
data['bmi'] = data['bmi'].fillna(data['bmi'].median())
data['smoker'] = data['smoker'].fillna(data['smoker'].mode()[0])
data['expenses'] = data['expenses'].fillna(data['expenses'].median())

data['region_northwest'] = data['region_northwest'].fillna(False)
data['region_southeast'] = data['region_southeast'].fillna(False)
data['region_southwest'] = data['region_southwest'].fillna(False)

data['region_northwest'] = data['region_northwest'].astype(int)
data['region_southeast'] = data['region_southeast'].astype(int)
data['region_southwest'] = data['region_southwest'].astype(int)
```

```python
import matplotlib.pyplot as plt

smokers_northwest = (data['region_northwest'] == 1) & (data['smoker'] == 1)
smokers_southeast = (data['region_southeast'] == 1) & (data['smoker'] == 1)
smokers_southwest = (data['region_southwest'] == 1) & (data['smoker'] == 1)

region_smoker_count = {
    'Northwest': smokers_northwest.sum(),
    'Southeast': smokers_southeast.sum(),
    'Southwest': smokers_southwest.sum()

}

regions = list(region_smoker_count.keys())
smokers = list(region_smoker_count.values())

plt.figure(figsize=(8, 6))
plt.bar(regions, smokers, color='skyblue', edgecolor='black')

plt.title('Number of Smokers in Each Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Number of Smokers', fontsize=12)

max_smokers_region = max(region_smoker_count, key=region_smoker_count.get)
max_smokers_count = region_smoker_count[max_smokers_region]
plt.text(max_smokers_region, max_smokers_count + 1, f'{max_smokers_count}',
ha='center', color='red', fontsize=12, fontweight='bold')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
insurance_data = pd.read_csv('/content/insurance.csv')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(insurance_data['age'], insurance_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0),
            color='blue', alpha=0.5, marker='o')
plt.title('Scatter Plot: Age vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking Status (1 = Yes, 0 = No)')
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7)
plt.show()
```

```python
filtered_data = insurance_data[(insurance_data['age'] >= 20) & (insurance_data['age'] <= 40)]
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0),
            color='green', alpha=0.6, marker='x')
plt.title('Scatter Plot: Age (20-40) vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking Status (1 = Yes, 0 = No)')
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.7)
plt.show()
```

```python
smoker_expenses = insurance_data.groupby('smoker')['expenses'].mean()
plt.figure(figsize=(8, 6))
smoker_expenses.plot(kind='bar', color=['green', 'blue'])
plt.title('Average Expenses: Smokers vs Non-Smokers')
plt.xlabel('Smoking Status')
plt.ylabel('Average Expenses')
plt.xticks(rotation=0)
plt.show()
```

```python
smokers_gender = insurance_data[insurance_data['smoker'] == 'yes'].groupby('sex').size()
plt.figure(figsize=(8, 6))
smokers_gender.plot(kind='bar', color=['pink', 'skyblue'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(insurance_data['age'], insurance_data['expenses'], color='blue', alpha=0.6)
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.show()
```

```python
teenage_smokers = insurance_data[(insurance_data['age'] >= 13) & (insurance_data['age'] <= 19) & (insurance_data['smoker'] == 'yes')]
teenage_smokers_count = teenage_smokers.shape[0]
print(f"Count of teenage smokers: {teenage_smokers_count}")
```
